```python
# 模拟生成一组实验数据
x = [1,3,5,7]
y = [0.41,0.38,0.43,0.46]
plt.figure(figsize=(8,8))
fig, ax = plt.subplots()
ax.plot(x, y, 'b--')
ax.set_xlabel('x')
ax.set_ylabel('y')

# 二次拟合
coef = np.polyfit(x, y, 2)
y_fit = np.polyval(coef, x)
ax.plot(x, y_fit, 'g')
# 找出其中的峰值/对称点
if coef[0] != 0:
    x0 = -0.5 * coef[1] / coef[0]            
    x0 = round(x0, 2)        
    ax.plot([x0]*5, np.linspace(min(y),max(y),5),'r--')
    print(x0)
else:
    raise ValueError('Fail to fit.')
plt.xlabel('放牧强度')
plt.ylabel('沙漠化指数SM')
plt.show()
coef
```
